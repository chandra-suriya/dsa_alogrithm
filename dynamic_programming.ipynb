```python
#Geek stones
def geek_steps(arr,ind,k,dp):
    if(ind==0):
        return 0
    
    
    if(dp[ind]!=-1):
        return dp[ind]
    
    min_dist = float('inf')
    for j in range(1,k+1):
        
        if(ind-j>=0):
            jumps = geek_steps(arr,ind-j,k,dp)+abs(arr[ind]-arr[ind-j])
            min_dist = min(min_dist,jumps)
            dp[ind] = min_dist
    return dp[ind]

k = 3
arr= [10, 30, 40, 50, 20]
ind=len(arr)

# dp with memoization
dp=[(-1)for i in range(0,ind)]

print(geek_steps(arr,ind-1,k,dp))
```

```python
# house robber problem(find the maximum sum of non adjacent elements in array)

# bruteforce 
def house_robber(nums,ind):
    if(ind<0):  #for handling the case (1-2)=-1
        return 0
    if(ind == 0):
        return nums[ind] #since we are entering the first element we should automatically sum
    
    pick = nums[ind] + house_robber(nums,ind-2) # picking the non adjacent element
    notpick = 0+house_robber(nums,ind-1) # not picking the lement and reduce the ind
    return(max(pick,notpick))

# dynamic programming(memoization,topdown approach)
def house_robber_dp(nums,ind,dp):
    #corner cases
    if(ind<0):
        return 0
    #base case
    if(ind == 0):
        return nums[ind]
    if(dp[ind]!=-1):
        return dp[ind]
    pick = nums[ind]+house_robber_dp(nums,ind-2,dp)
    notpick = 0+house_robber_dp(nums,ind-1,dp)
    dp[ind] = max(pick,notpick)
    return dp[ind]

# dynamic programming(tabulation,bottom up approach)

def top_down_house_robber(nums,ind,dp):
    #like base case
    dp[0]=nums[0]
    # neg = 0 #just for refernce
    for i in range(1,ind+1):
        take = nums[i]
        if(i>1): # just for handling negative values
            take+=dp[i-2] # picking the element
        nottake = 0+dp[i-1]# non picking the element
        dp[i] = max(take,nottake)

    return dp[ind]

# space optimisation
def space_optimisation(nums):
    n = len(nums)
    prev1 = nums[0] # like base case
    prev2 = 0  # like corner  case handling negative values
    curri=0
    for i in range(1,n):
        take = nums[i]
        if(i>1):
            take+=prev2
        nottake = 0+prev1
        curri = max(take,nottake)
        prev2 = prev1
        prev1 = curri
    return prev1

        


nums = [2,7,9,3,1]
n=len(nums)
dp=[(-1)]*n
print(house_robber(nums,n-1))
print(house_robber_dp(nums,n-1,dp))
print(top_down_house_robber(nums,n-1,dp))
print(space_optimisation(nums))
```

```python
# ninjas traning
def maxmise_tabulation_dp(arr,days,lasts,dp):
    dp[0][0] = max(arr[0][1],arr[0][2])
    dp[0][1] = max(arr[0][0],arr[0][2])
    dp[0][2] = max(arr[0][0],arr[0][1])
    dp[0][3] = max(arr[0][0],arr[0][1],arr[0][2])

    for day in range(1,days):
        for last in range(0,4):
            dp[day][last] = 0 # initial computation
            for task in range(0,3):
                if(task != last):
                    points = arr[day][task] + dp[day-1][task] # taking and not taking next iteration
                    dp[day][last] = max(points, dp[day][last])
    return dp[days-1][3]
n=3
arr = [[1,2,5],[3,1,1],[3,3,3]]
dp = [[-1]*(n+1) for i in range(0, len(arr))]
print(maxmise_tabulation_dp(arr,n,3,dp))
```

```python
## cheery pickup 3d grid
def cherry_pickup_3d(grid,i,j1,j2,r,c):
    ##base case
    if(j1<0 or j2<0 or j1>=c or j2>=c):
        return -100000008
    
    if(i==r-1): #entering into last row
        if(j1 == j2):# if it takes same column take any one either bob or alice
            return(grid[i][j1])
        else:
            return(grid[i][j1]+grid[i][j2])
    
    # compute all possible ways 
    # there are totally 9 possible including both alice bob to traverse
    
    maxi = 0
    for alice in range(-1,2):
        for bob in range(-1,2):
            # if both in same cell
            val=0
            if(j1 ==j2):
                val = grid[i][j1]
            else:
                val = grid[i][j1]+grid[i][j2]
            val +=cherry_pickup_3d(grid,i+1,j1+alice,j2+bob,r,c)
            maxi = max(maxi,val)
    return maxi


# memoization


def cherry_pickup_3dp(grid,i,j1,j2,r,c,dp):
    ##base case
    if(j1<0 or j2<0 or j1>=c or j2>=c):
        return int(-1e9)
    
    if(i==r-1): #entering into last row
        if(j1 == j2):# if it takes same column take any one either bob or alice
            return(grid[i][j1])
        else:
            return(grid[i][j1]+grid[i][j2])
    
    # check all the previous computed stored in dp
    if(dp[i][j1][j2] != -1):
        return dp[i][j1][j2]

    # compute all possible ways 
    # there are totally 9 possible including both alice bob to traverse
    
    maxi = 0
    for alice in range(-1,2):
        for bob in range(-1,2):
            # if both in same cell
            val=0
            if(j1 ==j2):
                val = grid[i][j1]
            else:
                val = grid[i][j1]+grid[i][j2]
            val +=cherry_pickup_3dp(grid,i+1,j1+alice,j2+bob,r,c,dp)
            maxi = max(maxi,val)
    dp[i][j1][j2] = maxi        
    return dp[i][j1][j2]


grid = [[2, 3, 1, 2], [3, 4, 2, 2], [5, 6, 3, 5]]
r,c = 3,4
dp = [[[-1 for j in range(c)] for i in range(c)] for k in range(r)]
print(cherry_pickup_3d(grid,0,0,c-1,r,c))
print(cherry_pickup_3dp(grid,0,0,c-1,r,c,dp))

#tabulation
#creating the base case

tdp = [[[0 for j in range(c)] for i in range(c)] for k in range(r)]
for alice in range(c):
    for bob in range(c):
        if(alice == bob):
            tdp[r-1][alice][bob] = grid[r-1][alice]
        else:
            tdp[r-1][alice][bob] = grid[r-1][alice]+grid[r-1][bob]

## normally traverseing
for i in range(r-2,-1,-1):
    for j1 in range(0,c):
        for j2 in range(0,c):
            
            maxi = -int(1e9)
            for alice in range(-1,2):
                for bob in range(-1,2):
                    # if both in same cell
                    val=0
                    if(alice ==bob):
                        val = grid[i][j1]
                    else:
                        val = grid[i][j1]+grid[i][j2]
                    if((j1+alice>=0 and j1+alice<c) and (j2+bob>=0 and j2+bob<c)):  
                        val +=tdp[i+1][j1+alice][j2+bob]
                    else:
                        val+=int(-1e9)
                    maxi = max(maxi,val)
            tdp[i][j1][j2] = maxi
print(tdp[0][0][c-1])    
    
```

```python
#unbound knapsack
def knapSack(val, wt,capacity):
        # code here
        n=len(wt)
        dp=[[0]*(capacity+1)for i in range(0,n)]
        # for base case
        #since for ind=0 ssine the weight might be different we have to 
        for i in range(0,capacity+1):
            #for infinite supplies we may add weight till the cacpacity less than weights
            dp[0][i] = (i//wt[0])*val[0]
        
        for ind in range(1,n):
            for wei in range(0,capacity+1):
                nottake = 0+dp[ind-1][wei]
                take = -1e9
                if(wt[ind]<=wei):
                    take = val[ind]+dp[ind][wei-wt[ind]]
                dp[ind][wei] = max(take,nottake)
                
        return dp[n-1][capacity]
        


val= [6, 1, 7, 7]
wt=[1, 3, 4, 5]
capacity = 8
print(knapSack(val, wt,capacity))
```
