```python
# generalised method to find the shortest destination from source to destination
import heapq

class Solution:
    def dijkstra(self, V, adj, S):
        # Priority queue to store (distance, node)
        pq = []
        # Initialize distances with infinity
        distTo = [float('inf')] * V
        # Distance to source is 0
        distTo[S] = 0
        heapq.heappush(pq, (0, S))
        
        while pq:
            # Get the node with the smallest distance
            dis, node = heapq.heappop(pq)
            
            # Traverse all adjacent nodes
            for it in adj[node]:
                v, w = it
                if dis + w < distTo[v]:
                    distTo[v] = dis + w
                    heapq.heappush(pq, (distTo[v], v))
        
        # Return the shortest distances from source
        return distTo


if __name__ == "__main__":
    # Driver code
    V,S = 6,2
    data = [[0,1,1],[0,3,3],[1,3,1],[1,2,2],[3,2,1],[3,4,2],[4,5,1],[5,2,2]]
    adj = [[] for _ in range(V)]
    for val in data:
        adj[val[0]].append([val[1],val[2]])
        adj[val[1]].append([val[0],val[2]])
    #print(adj)
    
    obj = Solution()
    res = obj.dijkstra(V, adj, S)
    
    print(" ".join(map(str, res)))
```

```python
## accounts merge DSU
class DisjointSet():
    def __init__(self,n):
        self.parent = [i for i in range(n+1)]
        self.rank = [0]*(n+1)  # initially rank was zero for everything
        self.size = [1] *(n+1) #initally each node was a samenode having the size 1

        
    
    def getparent(self, node):
        if(node == self.parent[node]):
            return node
        self.parent[node] = self.getparent(self.parent[node])
        return self.parent[node]
     
    def union_byrank(self, u,v):
        ulp_u = self.getparent(u)
        ulp_v = self.getparent(v)

        #checking if having the same parent
        if(ulp_u == ulp_v):
            return

        # checking the rank of the node based on the condition
        
        if(self.rank[ulp_u] < self.rank[ulp_v]):
            # since the rank of the parent of u node is grater than the parent of  node
            # we might want to add the biggest parent node to the samlles tparent node
            self.parent[ulp_u] = ulp_v #adding the parent node to 

        elif(self.rank[ulp_u] >  self.rank[ulp_v]):

            # like above the alternate conditon 
            self.parent[ulp_v] = ulp_u

        else:
            self.parent[ulp_u] = ulp_v # adding any one node to any othe parent node
            self.rank[ulp_v]+=1

accounts = [["John","johnsmith@mail.com","john_newyork@mail.com"],
            ["John","johnsmith@mail.com","john00@mail.com"],
            ["Mary","mary@mail.com"],["John","johnnybravo@mail.com"]]

mailind = dict()
n = len(accounts)
accounts.sort()
ds = DisjointSet(n)
for val in range(0,len(accounts)):
    for ind in range(1,len(accounts[val])):
        mail = accounts[val][ind]
        if(mail in mailind.values()):
            ds.union_byrank(val,mailind[accounts[val][ind]])
        else:
            mailind[mail]=val

mergedMail = [[] for _ in range(n)]
for mail,ind in mailind.items():
    node = ds.getparent(ind)
    mergedMail[node].append(mail)


ans = []
for i in range(n):
    if not mergedMail[i]:
        continue
    mergedMail[i].sort()
    temp = []
    temp.append(accounts[i][0])
    for val in mergedMail[i]:
        temp.append(val)
    ans.append(temp)
ans.sort()
print(ans)
```
